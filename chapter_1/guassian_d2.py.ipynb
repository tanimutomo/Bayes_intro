```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def gauss(x, mu, sig2):
    return np.exp(- (x - mu)**2 / 2 * sig2) / np.sqrt(2 * np.pi * sig2)
```

```python
x = np.linspace(0, 2 * math.pi, 30) # 配列
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x1 = 10 * np.random.rand(100)
y1 = np.sin(x1)
```

```python
plt.scatter(x1,y1)
plt.show()
```

```python
x2 = 10 * np.random.rand(1000) - 5
# x2_mean = np.mean(x2)
# x2_var = np.var(x2)
y2 = gauss(x2, 0, 1)
```

```python
plt.scatter(x2,y2)
plt.show()
```
